```python
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

# Create HTML request from wikipedia URL

```python
wikiurl="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

# Parse data from the html into a beautifulsoup object

```python
soup = BeautifulSoup(response.text, 'html.parser')
torontotable=soup.find('table',{'class':"wikitable"})
```

```python
print(torontotable) #confirm correct output
```

# Convert list to dataframe

```python
df=pd.read_html(str(torontotable))
df=pd.DataFrame(df[0])
df.head()
```

# Drop all boroughs that are 'Not assigned'

```python
df = df[df['Borough'] != 'Not assigned']
df.head(40)
```

# Combine neighborhoods with same postal area into one row with neighborhoods separated with comma

```python
df_temp = pd.DataFrame({'Postal Code':df['Postal Code'].unique()})
df_temp['Borough']=pd.DataFrame(list(set(df['Borough'].loc[df['Postal Code'] == x['Postal Code']])) for i, x in df_temp.iterrows())
df_temp['Neighbourhood']=pd.Series(list(set(df['Neighbourhood'].loc[df['Postal Code'] == x['Postal Code']])) for i, x in df_temp.iterrows())
df_temp['Neighbourhood']=df_temp['Neighbourhood'].apply(lambda x: ', '.join(x))
```

# Combine rows with Not assigned assigned neighborhood, but assigned borough

```python
df[['Neighbourhood']].replace('Not assigned', df[['Borough']], inplace=True)
df.head(20)
```
